```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Data/sim_g.csv')
x = np.array(data.iloc[:,1:data.shape[1]])
y = np.array(data.iloc[:,0])
```

```python
## with interaction but no useless predictors
model1 = Sequential()
model1.add(Dense(20, input_dim=np.shape(x)[1]-5, activation='relu'))
model1.add(Dense(10, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))

model1.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

trace1 = model1.fit(x[0:2500,0:12], y[0:2500],epochs=50,batch_size=64)
score1 = model1.evaluate(x[2500:3000,0:12], y[2500:3000], batch_size=64)
```

```python
score1
```

```python
## without interaction or useless predictor
model2 = Sequential()
model2.add(Dense(20, input_dim=np.shape(x)[1]-10, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))

model2.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

trace2 = model2.fit(x[0:2500,0:7], y[0:2500],epochs=50,batch_size=64)
score2 = model2.evaluate(x[2500:3000,0:7], y[2500:3000], batch_size=64)
```

```python
score2
```

```python
## with interaction and useless predictor
model3 = Sequential()
model3.add(Dense(20, input_dim=np.shape(x)[1], activation='relu'))
model3.add(Dense(10, activation='relu'))
model3.add(Dense(1, activation='sigmoid'))

model3.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

trace3 = model3.fit(x[0:2500,:], y[0:2500],epochs=50,batch_size=64)
score3 = model3.evaluate(x[2500:3000,:], y[2500:3000], batch_size=64)
```

```python
score3
```

```python
plt.title('2 hidden layers, narrow')
plt.plot(trace1.history['acc'],label='interaction, no garbage')
plt.plot(trace2.history['acc'],label='no interaction, no garbage')
plt.plot(trace3.history['acc'],label='interaction and garbage')
plt.legend(loc=4)
plt.show()
```

```python
##########################################################################################################
### Set up workspace
##########################################################################################################
# Import libraries - general
import numpy as np
import pylab as pl
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../Data/train.csv', keep_default_na=False,  na_values='-1')
train.shape
```

```python
train.head(10)
```

```python
KNNmat0=train.drop(train.columns[[0, 1]], axis=1)
KNNmat0.head(5)
```

```python
KNNmat_complete=KNNmat0.dropna(axis=0, how='any')
```

```python
KNNmat_complete.shape
```

```python
KNNmat_complete.head(10)
```

```python
complete_index=np.array(KNNmat_complete.index)
complete_index
```

```python
miss_index=set(range(595212))-set(complete_index)
miss_index=np.array(list(miss_index),dtype='int64')
miss_index
```

```python
KNNmat1 = KNNmat0.replace(-1, np.nan, regex=True)
```

```python
KNNmat1.mean(axis=0)
```

```python
KNNmat2=KNNmat1.fillna(KNNmat1.mean(axis=0))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
KNNmat3 = pd.DataFrame(scaler.fit_transform(KNNmat2), columns=KNNmat2.columns)
```

```python
KNNmat3.head(5)
```

```python
from sklearn.neighbors import NearestNeighbors
nbrs = NearestNeighbors(n_neighbors=1, algorithm='auto').fit(KNNmat3.loc[complete_index])
```

```python
indices = nbrs.kneighbors(KNNmat3.loc[miss_index],return_distance=False)
```

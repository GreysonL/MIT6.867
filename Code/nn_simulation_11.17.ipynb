```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../Data/sim_g.csv')
x = np.array(data.iloc[:,1:data.shape[1]])
y = np.array(data.iloc[:,0])
```

```python
## with interaction
model1 = Sequential()
model1.add(Dense(20, input_dim=np.shape(x)[1], activation='relu'))
model1.add(Dense(10, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))

model1.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model1.fit(x[0:2500,:], y[0:2500],epochs=50,batch_size=64)
score1 = model1.evaluate(x[2500:3000,:], y[2500:3000], batch_size=64)
```

```python
score1
```

```python
## without interaction
model2 = Sequential()
model2.add(Dense(20, input_dim=np.shape(x)[1]-5, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))

model2.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model2.fit(x[0:2500,0:7], y[0:2500],epochs=50,batch_size=64)
score2 = model2.evaluate(x[2500:3000,0:7], y[2500:3000], batch_size=64)
```
